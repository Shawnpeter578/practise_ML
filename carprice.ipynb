```python
import pandas as pd
```

```python
df = pd.read_csv('/content/carprices.csv')
df
```

```python
dum = pd.get_dummies(df['Car Model'])
dum
```

```python
merge = pd.concat([df,dum],axis='columns')

merge
```

```python
final = merge.drop(["Car Model", "Mercedez Benz C class"], axis='columns')
final
```

```python
X = final.drop(['Sell Price($)'],axis='columns')
X
```

```python
y = final['Sell Price($)']
y
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(X,y)
reg
```

```python
reg.predict([[8600,7,0,1]])
```

```python
reg.score(X,y)
```
