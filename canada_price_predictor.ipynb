```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("/content/canada_price.csv")
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('price')
plt.scatter(df.year,df["per capita income (US$)"],color='red',marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["year"]], df["per capita income (US$)"])
```

```python
reg.predict([[2026]])
```

```python
m = reg.coef_
```

```python
b = reg.intercept_
```

```python
price = m * 2020 + b
price
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('price')
plt.scatter(df.year,df["per capita income (US$)"],color='red',marker='+')
plt.plot(df.year,reg.predict(df[["year"]]),color='blue')
```
